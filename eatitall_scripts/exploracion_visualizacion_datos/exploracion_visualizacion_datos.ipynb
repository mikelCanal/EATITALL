```python
def histograma(nombre_variable,df):
    # Crear el histograma
    plt.hist(df[nombre_variable], bins='auto', alpha=0.7, color='blue', edgecolor='black')

    # Agregar títulos y etiquetas
    plt.title('Histograma de %s'%nombre_variable)
    plt.xlabel('Valores')
    plt.ylabel('Frecuencia')

    # Mostrar el histograma
    plt.show()
```

```python
def histogramas_3x3(nombres_variables, df):
    # Crear una figura con subplots en una matriz 3x3
    fig, axes = plt.subplots(3, 3, figsize=(15, 10))
    axes = axes.ravel() # Convertir la matriz de ejes en un array 1D para facilitar la iteración

    # Iterar sobre las variables y los ejes para crear los histogramas
    for i, nombre_variable in enumerate(nombres_variables):
        # Crear el histograma en el eje correspondiente
        axes[i].hist(df[nombre_variable], bins='auto', alpha=0.7, color='blue', edgecolor='black')

        # Agregar títulos y etiquetas
        axes[i].set_title('Histograma de %s' % nombre_variable)
        axes[i].set_xlabel('Valores')
        axes[i].set_ylabel('Frecuencia')

    # Ajustar el layout para evitar la superposición
    plt.tight_layout()

    # Mostrar la figura con todos los histogramas
    plt.show()

nombres_variables=['edad', 'acv', 'hba1c', 'fg', 'tg','talla','peso','tas','tad']
histogramas_3x3(nombres_variables, df)
```
