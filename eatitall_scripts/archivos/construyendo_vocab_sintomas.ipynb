```python
# Dividir el string en líneas y luego cada línea en partes
lineas = datos.strip().split("\n")
datos_json = []

for linea in lineas:
    partes = linea.split("\t")  # Asumiendo que el separador es una tabulación
    objeto_json = {
        "descripción": partes[0],
        "env. accidental": partes[1],
        "env.autolitico": partes[2],
        "env. agresión": partes[3],
        "env. indeterminado": partes[4],
        "efectos adversos": None if partes[5] == "-" else partes[5],
        "infradosificación": None if partes[6] == "-" else partes[6],
    }
    datos_json.append(objeto_json)

# Si necesitas convertir la lista a un string JSON
import json
json_str = json.dumps(datos_json, ensure_ascii=False, indent=2)
print(json_str)

with open("vocab_farmacos_y_productos_quimicos.json", "w", encoding="utf-8") as archivo_json:
    archivo_json.write(json_str)
```
