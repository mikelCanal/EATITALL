```python
import json

# Cargar el archivo JSON
with open('./../../archivos/vocab_farmacos_y_productos_quimicos.json', 'r') as file:
    data = json.load(file)

# Extraer los nombres
env_acc = [entry['env. accidental'] for entry in data['farmacos y productos químicos']]
env_autolitico=[entry['env.autolitico'] for entry in data['farmacos y productos químicos']]
env_agresion=[entry['env. agresión'] for entry in data['farmacos y productos químicos']]
env_indeterminado=[entry['env. indeterminado'] for entry in data['farmacos y productos químicos']]
```
