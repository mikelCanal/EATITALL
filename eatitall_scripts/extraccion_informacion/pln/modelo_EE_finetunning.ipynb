```python
import sys

sys.path.insert(1, '/home/eatitall_scripts')
sys.path.insert(1, '/home/root/pctobs/lib/python3.8/site-packages')
import json

# Cargar el archivo JSON
with open('./../../archivos/vocab_farmacos_y_productos_quimicos.json', 'r') as file:
    data = json.load(file)

# Extraer los nombres
nombres = [entry['nombre'] for entry in data['farmacos y productos químicos']]
```

```python
nombres_sel=nombres[:256]
nombres_sel
```

```python
import random

# Plantillas de frases ejemplo
plantillas = [
    "Se recetó {} para tratar la infección bacteriana.",
    "El paciente fue tratado con {} tras su diagnóstico de hipertensión.",
    "Durante la consulta, se sugirió añadir {} al régimen terapéutico.",
    "Se ha reportado una reacción adversa grave al administrar {}.",
    "Se debe ajustar la dosis de {} debido a los efectos secundarios observados.",
    "En la revisión de su tratamiento, se recomendó sustituir su medicación actual por {}."
]

# Generar frases usando las plantillas
datos_entrenamiento = []
datos_validacion = []
for nombre in nombres_sel:
    plantilla = random.choice(plantillas)
    frase = plantilla.format(nombre)
    start_index = frase.index(nombre)
    end_index = start_index + len(nombre)
    datos_entrenamiento.append((frase, {'entities': [(start_index, end_index, nombre)]}))
    #### Validación
    plantilla = random.choice(plantillas)
    frase = plantilla.format(nombre)
    start_index = frase.index(nombre)
    end_index = start_index + len(nombre)
    datos_validacion.append((frase, {'entities': [(start_index, end_index, nombre)]}))
```

```python
import csv

# Supongamos que datos_entrenamiento contiene frases y sus anotaciones
# datos_entrenamiento = [('Frase 1', {'entities': [(inicio, fin, 'Tipo')]}), ...]

# Abrir un archivo CSV para escribir
with open('datos_entrenamiento.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    
    # Escribir los títulos de las columnas
    writer.writerow(['Texto', 'Entidades'])
    
    # Escribir los datos
    for texto, anotacion in datos_entrenamiento:
        # Convertir la lista de entidades a una cadena para guardarla en CSV
        entidades = ';'.join([f"({label},{start},{end})" for start, end, label in anotacion['entities']])
        writer.writerow([texto, entidades])

with open('datos_validacion.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    
    # Escribir los títulos de las columnas
    writer.writerow(['Texto', 'Entidades'])
    
    # Escribir los datos
    for texto, anotacion in datos_validacion:
        # Convertir la lista de entidades a una cadena para guardarla en CSV
        entidades = ';'.join([f"({label},{start},{end}" for start, end, label in anotacion['entities']])
        writer.writerow([texto, entidades])
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification, Trainer, TrainingArguments, TFAutoModelForTokenClassification  
from datasets import load_dataset, load_metric
import numpy as np

# Cargar el tokenizer y el modelo
tokenizer = AutoTokenizer.from_pretrained("PlanTL-GOB-ES/bsc-bio-ehr-es")
model = TFAutoModelForTokenClassification  .from_pretrained("PlanTL-GOB-ES/bsc-bio-ehr-es", num_labels=10, from_pt=True)

# Cargar los datasets
dataset = load_dataset('csv', data_files={'train': 'datos_entrenamiento.csv', 'validation': 'datos_validacion.csv'})

# Cargar la métrica
metric = load_metric('seqeval')

label_map = {}
for i, nombre in enumerate(nombres_sel, start=0):
    label_map[nombre] = i

# Función para procesar los datos
def tokenize_and_align_labels(examples):
    # Tokeniza el texto. Asegúrate de que 'Texto' es el nombre correcto de la columna.
    tokenized_inputs = tokenizer(examples['Texto'], truncation=True, padding=True, is_split_into_words=True, max_length=128, return_offsets_mapping=True)
    
    labels = []
    # Debug: imprimir el tamaño del lote de tokenización y de etiquetas para asegurar alineación
    print("Batch size:", len(tokenized_inputs['input_ids']), "Labels size:", len(examples['Entidades']))

    for i, label_list in enumerate(examples['Entidades']):
        print("i: ",i)
        # Obtén los word_ids y offset mappings para cada entrada en el lote
        word_ids = tokenized_inputs.word_ids(batch_index=i)
        offset_mapping = tokenized_inputs['offset_mapping'][i]
        label_ids = [-100] * len(word_ids)  # Usar -100 para ignorar tokens en el cálculo de la pérdida
        print("hola1")
        # Recorre cada entidad y asigna las etiquetas
        
        print("label_list: ",label_list)
        label_list_clean = label_list.strip('()').split(',')
        label_type = label_list_clean[0]
        start_char = int(label_list_clean[1])
        end_char = int(label_list_clean[2])
        # Convierte la etiqueta a un índice numérico, si es necesario
        print("hola2")
        label_index = label_map[label_type]
        print("hola3")

        # Asignar la etiqueta correcta a los tokens correspondientes
        print("offset_mapping",offset_mapping)
        for idx, (offset_start, offset_end) in enumerate(offset_mapping):
            if start_char < offset_end and end_char > offset_start:  # El token está dentro del rango de la entidad
                if start_char <= offset_start:  # Comienza la entidad
                    label_ids[idx] = label_index  # Aquí podrías usar diferentes etiquetas para B/I si estás usando BIO

        labels.append(label_ids)

    tokenized_inputs['labels'] = labels
    return tokenized_inputs


tokenized_datasets = dataset.map(tokenize_and_align_labels, batched=True)

# Función para calcular las métricas durante la evaluación
def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=2)
    
    # Extraer las etiquetas y predicciones reales
    true_labels = [[label_list[l] for l in label_line if l != -100] for label_line in labels]
    true_predictions = [
        [label_list[p] for (p, l) in zip(prediction_line, label_line) if l != -100]
        for prediction_line, label_line in zip(predictions, labels)
    ]
    
    # Calcular las métricas usando seqeval
    results = metric.compute(predictions=true_predictions, references=true_labels)
    return {
        "precision": results['overall_precision'],
        "recall": results['overall_recall'],
        "f1": results['overall_f1'],
        "accuracy": results['overall_accuracy'],
    }

# Configuración del Trainer
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['validation'],
    compute_metrics=compute_metrics
)

# Entrenar el modelo
trainer.train()
```
